```python
import pandas as pd
import numpy as np

# Create DataFrame
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}

energy_df = pd.DataFrame(data)

# Create copies
energy_df_mean = energy_df.copy()
energy_df_ffill = energy_df.copy()
energy_df_bfill = energy_df.copy()

# ---------------- Mean Imputation ----------------
energy_df_mean["Energy Consumption (MWh)"] = energy_df_mean["Energy Consumption (MWh)"].fillna(
    energy_df_mean["Energy Consumption (MWh)"].mean()
)

energy_df_mean["Cost (Million $)"] = energy_df_mean["Cost (Million $)"].fillna(
    energy_df_mean["Cost (Million $)"].mean()
)

print("Data After Mean Imputation:")
print(energy_df_mean)

# ---------------- Forward Fill ----------------
print("\nBefore Forward Fill:")
print(energy_df_ffill)

forward_filled_df = energy_df_ffill.ffill()

print("\nData After Forward Fill:")
print(forward_filled_df)

# ---------------- Backward Fill ----------------
print("\nBefore Backward Fill:")
print(energy_df_bfill)

backward_filled_df = energy_df_bfill.bfill()

print("\nData After Backward Fill:")
print(backward_filled_df)
```

```python
# Green Skilling & Sustainability Dataset Analysis
import pandas as pd
df = pd.read_csv("GreenSkilling_Sustainability_Dataset.csv")
```

```python
print("First 5 Rows")
print(df.head())
```

```python
print("\nDataset Shape")
print(df.shape)
```

```python
print("\nColumn Names")
print(df.columns)
```

```python
print("\nDataset Information")
print(df.info())
```

```python
print("\nStatistical Summary")
print(df.describe())
```

```python
print("\nNull Values")
print(df.isnull().sum())
```

```python
# Numerical columns are filled with their mean values.

df["Solar_Energy_%"] = df["Solar_Energy_%"].fillna(df["Solar_Energy_%"].mean())

df["Green_Skilling_Training"] = df["Green_Skilling_Training"].fillna(
    df["Green_Skilling_Training"].mean()
)

print("\nNull Values After Handling")
print(df.isnull().sum())
```

```python
# Correlation

print("\nCorrelation Matrix")
print(df.corr(numeric_only=True))
```

```python
print("Average Solar Energy")
print(df["Solar_Energy_%"].mean())
```
